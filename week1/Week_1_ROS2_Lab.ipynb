# 2. ROS 2 Conceptual Overview

ROS 2 is a framework for building robot software using a distributed architecture.

## 2.1 Key vocabulary
- Node: a process that performs computation (example: controller, sensor driver, planner)
- Topic: a named communication channel for streaming messages
- Publisher: sends messages on a topic
- Subscriber: receives messages from a topic
- Service: request/response communication (like calling a function over the network)
- Action: long-running goal with feedback (used for navigation goals)
- Package: a folder with ROS 2 code, dependencies, and build information
- Workspace: a folder that contains one or more ROS 2 packages and build outputs

## 2.2 Conceptual diagram (topics)
Below is a conceptual diagram showing topic communication.

```
+-----------------------+                     +-----------------------+
|       Node A          |                     |        Node B         |
|  (Publisher)          |                     |   (Subscriber)        |
|                       |                     |                       |
| publishes Twist msgs  |   /cmd_vel topic    | receives Twist msgs   |
| at 10 Hz              +-------------------->| and acts on them      |
+-----------------------+                     +-----------------------+
```

## 2.3 How ROS 2 runs your code
When you type:
`ros2 run <package> <executable>`

ROS 2:
1. Finds the installed package in your sourced workspace
2. Locates the registered executable (entry point)
3. Runs it as a normal process
4. That process becomes a node when it calls `rclpy.init()` and constructs a `Node` object

# 4. Create a ROS 2 Workspace

A workspace is a folder where you build and install packages. 

## 4.1 Create the standard structure
```bash
mkdir -p ~/ros2_ws/src
cd ~/ros2_ws
```

## 4.2 First build (empty workspace)
```bash
colcon build
```

## 4.3 Source the workspace (so ROS 2 can find built packages)
```bash
source install/setup.bash
```

## 4.4 Make sourcing persistent (recommended)
This automatically sources your workspace when you open a new terminal:
```bash
echo "source /opt/ros/humble/setup.bash" >> ~/.bashrc
echo "source ~/ros2_ws/install/setup.bash" >> ~/.bashrc
```
Then either open a new terminal or run:
```bash
source ~/.bashrc
```

## 4.5 Workspace folder meanings
After building, you will have:

- `src/`     : your packages (source code)
- `build/`   : build artifacts (intermediate files)
- `install/` : installed packages and executables
- `log/`     : logs from the build process

Quick view:
```bash
cd ~/ros2_ws
ls
```

# 7. Write a First Node (Python)

Goal: Create a node that prints a message when it starts.

## 7.1 Create the node file
Create:
`~/ros2_ws/src/my_first_pkg/my_first_pkg/simple_node.py`

You may use `nano`:
```bash
nano ~/ros2_ws/src/my_first_pkg/my_first_pkg/simple_node.py
```

Paste the following code in the editor.


````python

import rclpy
from rclpy.node import Node


class SimpleNode(Node):
    def __init__(self):
        super().__init__('simple_node')
        self.get_logger().info('Hello, ROS 2')


def main(args=None):
    rclpy.init(args=args)
    node = SimpleNode()

    # spin_once lets us create the node, log once, and exit cleanly
    rclpy.spin_once(node, timeout_sec=0.1)

    node.destroy_node()
    rclpy.shutdown()


if __name__ == '__main__':
    main()
````

## 7.2 Make file executable (good practice)
```bash
chmod +x ~/ros2_ws/src/my_first_pkg/my_first_pkg/simple_node.py
```

# 8. Register the Node as an Executable (Entry Point)

ROS 2 needs to know what command name maps to which Python function.

## 8.1 Edit setup.py
Open:
```bash
nano ~/ros2_ws/src/my_first_pkg/setup.py
```

Find the `entry_points` section and make it look like this (only the console_scripts part matters):

```python
entry_points={
    'console_scripts': [
        'simple_node = my_first_pkg.simple_node:main',
    ],
},
```

Save the file.

## 8.2 Rebuild and source
```bash
cd ~/ros2_ws
colcon build
source install/setup.bash
```

# 10. Troubleshooting (Read This Before Asking for Help)

## Problem A: `ros2: command not found`
Cause: ROS 2 is not sourced in the terminal.
Fix:
```bash
source /opt/ros/humble/setup.bash
```
Then retry:
```bash
ros2 --version
```

## Problem B: `colcon: command not found`
Cause: colcon not installed or not in PATH.
Fix (Ubuntu):
```bash
sudo apt update
sudo apt install python3-colcon-common-extensions
```

## Problem C: `Package 'my_first_pkg' not found`
Common causes:
1. Workspace not sourced
2. Build failed
3. Package created in the wrong folder

Fix steps:
```bash
cd ~/ros2_ws
colcon build
source install/setup.bash
ros2 pkg list | grep my_first_pkg
```
Also confirm you created it inside `~/ros2_ws/src`.

## Problem D: `ros2 run my_first_pkg simple_node` says executable not found
Cause: entry point not registered correctly or build not re-run.
Fix:
1. Re-check `setup.py` console_scripts line
2. Rebuild and source:
```bash
cd ~/ros2_ws
colcon build
source install/setup.bash
```

## Problem E: Python error (SyntaxError, ModuleNotFoundError)
Cause: code typo or missing dependency.
Fix:
- Open the file and check indentation and quotes carefully
- Ensure `import rclpy` works (ROS environment sourced)
- Rebuild after changes

## Problem F: You edited files but changes do not take effect
Cause: You forgot to rebuild and/or source.
Fix:
```bash
cd ~/ros2_ws
colcon build
source install/setup.bash
```

## How to ask for help (required format)
When you ask for help, include:
1. The exact command you ran
2. The full error text (copy/paste)
3. Your folder structure output:
```bash
cd ~/ros2_ws
find src/my_first_pkg -maxdepth 2 -type f
```
